```python
#1- Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
#   nominal, ordinal, interval, and ratio scales.

# statistics, data can be classified into two main types: qualitative (also called categorical) and quantitative
# (also called numerical). These types of data are further divided based on the level of measurement they represent.
#  Below, I explain each type of data and the four different measurement scales: nominal, ordinal, interval, and ratio.

# Qualitative Data (Categorical Data)---
# Qualitative data consists of information that describes categories or qualities. This data does not involve numbers 
# or quantities but rather focuses on characteristics. Qualitative data can be further classified into two types:

# Nominal Data---   Nominal data represents categories or labels with no inherent order or ranking.
# Example------
# Gender (Male, Female)
# Hair color (Black, Blonde, Brown, Red)
# Marital status (Single, Married, Divorced)
# Key characteristic: The values are merely names or labels, and there is no meaningful way to order them.

# Ordinal Data---   Ordinal data represents categories with a meaningful order or ranking, but the differences
# between the ranks are not uniform or known.
# Example------
# Education level (High School, Bachelor's, Master's, PhD)
# Likert scale ratings (Strongly Agree, Agree, Neutral, Disagree, Strongly Disagree)
# Ranks in a competition (1st place, 2nd place, 3rd place)
# Key characteristic: The categories can be ordered, but the exact difference between the ranks is not measurable or consistent.
# Quantitative Data (Numerical Data)
# Quantitative data consists of numerical values that can be measured or counted. This type of data can be further
# classified into two types based on the scale of measurement:

# Interval Data---  Interval data consists of numerical values with meaningful differences between them, but there is no true zero point.
# Example-----
# Temperature in Celsius or Fahrenheit (e.g., 10°C, 20°C, 30°C)
# IQ scores (e.g., 85, 100, 120)
# Key characteristic: The differences between values are meaningful (e.g., a difference of 10°C is the same at any
# point), but there is no absolute zero. In the case of temperature, 0°C does not represent "no temperature,
# " it just represents the freezing point of water.

# Ratio Data---   Ratio data consists of numerical values with both meaningful differences between values and a true zero point, which allows for the calculation of ratios.
# Example-----
# Height (e.g., 0 cm, 150 cm, 200 cm)
# Weight (e.g., 0 kg, 50 kg, 100 kg)
# Income (e.g., $0, $50,000, $100,000)
# Key characteristic: In addition to meaningful differences, ratio data has an absolute zero point, meaning "zero" 
# represents a complete absence of the quantity. This allows for the full range of mathematical operations, including 
# ratios (e.g., a person who weighs 100 kg weighs twice as much as a person who weighs 50 kg).




#2- What are the measures of central tendency, and when should you use each? Discuss the mean, median,and mode with 
#   examples and situations where each is appropriate.

#  Measures of central tendency are statistical tools that summarize a set of data by identifying the central point or typical value around which the data tends to cluster. The three most common measures of central tendency are:

#  Mean
#  Median
#  Mode

#1  Mean (Arithmetic Average)----- The mean is the sum of all the values in a dataset divided by the number of values.
#  It is the most commonly used measure of central tendency.
 
# Example:
# Consider the following dataset representing the ages of 5 people: 
# 10,15,20,25,30
# The mean is calculated as----
# Mean=10+15+20+25+30 / 5 = 100 / 5 = 20

# So, the mean age is 20 years.

# When to Use---

# Best for Symmetrical Data: The mean is most appropriate when the data distribution is symmetrical or approximately
# normal (i.e., the data is evenly spread out around a central point).
# When All Data Points Are Important: The mean takes into account every value in the dataset, making it useful when 
# each data point carries equal importance (e.g., calculating the average test score of students).




#2  Median--  The median is the middle value of a dataset when the values are arranged in ascending or descending order.
#  If the dataset has an even number of values, the median is the average of the two middle values.

# Example:
# Consider the following dataset representing the ages of 5 people---  10,15,20,25,30

# The median is the middle value, which in this case is:   median = 20
# If the dataset had an even number of values, say: 10,15,20,25 The median would be--
# median = 15 + 20 / 2 = 17.5

# When to Use:
# Best for Skewed Data: The median is ideal for data that is skewed (not symmetrically distributed) or when there are
# outliers. For example, in income data, where a few very high incomes could distort the mean, the median gives a 
# better representation of the typical income.
# Resistant to Outliers: Unlike the mean, the median is not affected by outliers. For instance, if one person in a
# dataset earns $1 million while others earn $50,000, the median will still reflect the central tendency better than
# the mean.




#3 Mode--  The mode is the value that appears most frequently in a dataset. A dataset may have one mode (unimodal),
#  two modes (bimodal), or more modes (multimodal). If all values appear with the same frequency, the dataset is said
#  to have no mode.

# Example:
# Consider the following dataset of test scores:  50,60,60,70,80,80,90

# The mode is-  Mode = 60 and 80
# Since both 60 and 80 appear twice, they are the modes of the dataset (this is a bimodal distribution).

# If the dataset were: 50,60,70,80,90 The mode would be-   Mode=No mode (since all values appear only once).

# When to Use--

# Best for Categorical Data: The mode is useful when dealing with categorical data, where we want to know the most 
# common category or value. For example, determining the most common color of cars sold in a dealership.
# Useful for Discrete Data: For datasets where the values are discrete (e.g., the number of children in a family), 
# the mode can give a clear indication of the most frequent value.
# For Data with Multiple Peaks: When a dataset has multiple frequent values (multimodal), the mode can help identify
# those peaks.




#3- Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

#  Concept of Dispersion---
#  In statistics, dispersion refers to the degree to which data values spread out from the central value 
#  (such as the mean, median, or mode). Dispersion gives us an idea of how much variability or "spread" exists within
#  a dataset. While measures of central tendency (like the mean and median) tell us where the data tends to center, 
#  measures of dispersion tell us how far the data points are from that central point.

#  Understanding dispersion is crucial because two datasets with the same mean can still have very different
#  distributions—one might be tightly clustered around the mean, while the other could be more spread out. 

#  Key Measures of Dispersion--
#  The most common measures of dispersion are variance and standard deviation. Both of these measures quantify how 
#  much the values in a dataset differ from the mean, but they do so in different ways.

#1.Variance---- Variance measures the average squared deviation from the mean. It quantifies how much the data points
#  vary from the mean, but it is expressed in squared units of the original data, which can make it harder to interpret directly.

#  When to Use---
#  Variance is typically used when dealing with the spread of data in a mathematical or theoretical context.
#  It is particularly useful in inferential statistics and in calculating other metrics, like standard deviation and correlation. 

#2.Standard Deviation----  The standard deviation is simply the square root of the variance. It measures the average
#  amount by which data points deviate from the mean, and it is expressed in the same units as the original data.
#  This makes the standard deviation more intuitive and easier to interpret than the variance.

# When to Use---
# Standard deviation is more commonly used in practice because it is in the same units as the data, making it easier to interpret.
# Standard deviation is especially useful for making comparisons between different datasets.



#4- What is a box plot, and what can it tell you about the distribution of data?

#  A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset. 
#  It provides a visual summary of the central tendency, dispersion, and skewness of the data, along with the presence
#  of outliers. Box plots are particularly useful for comparing distributions between multiple datasets

#  What a Box Plot Can Tell You About the Distribution of Data-----
#  A box plot gives you several insights about the dataset:

#1. Central Tendency (Median)----

#   The median (Q2) is shown as the line inside the box. It represents the middle value of the data and indicates the
#   central tendency of the dataset.
#   If the median is near the center of the box, the data is likely symmetrically distributed. If the median is 
#   skewed toward one side, the data may be skewed in that direction.

#2. Spread of the Data (IQR and Range)----

#   The box itself represents the IQR (the middle 50% of the data), showing the spread between the first (Q1) and 
#   third quartile (Q3). A wider box indicates more variability, while a narrower box suggests the data is more tightly clustered.
#   The whiskers show the range of the data, excluding outliers. A longer whisker indicates a greater spread, 
#   while a shorter whisker suggests a narrower distribution.

#3. Skewness----

#   Skewness refers to the asymmetry of the distribution.
#   If the median is closer to the top of the box, and the upper whisker is longer than the lower one, the data is
#   positively skewed (skewed to the right).
#   If the median is closer to the bottom of the box, and the lower whisker is longer than the upper one, the data
#   is negatively skewed (skewed to the left).
#   If the median is in the middle of the box, and the whiskers are approximately equal in length, the data is 
#   symmetrical or approximately normal.

#4. Outliers----

#   Outliers are points that lie outside the whiskers, often defined as points more than 1.5 times the IQR away
#   from Q1 or Q3. These are marked individually on the plot.
#   Outliers can indicate rare or extreme values that may require further investigation. For example, in income data,
#   outliers may represent exceptionally high earners.

#5. Comparing Multiple Datasets----

#   Box plots are especially useful when comparing the distributions of multiple datasets side by side. You can see:
#-   The central tendency of each dataset (median line).
#-   The spread of each dataset (IQR and range).
#-   The presence of outliers.




#5- Discuss the role of random sampling in making inferences about populations.

#  Role of Random Sampling in Making Inferences About Populations-----
#  Random sampling is a fundamental concept in statistics that plays a crucial role in making valid inferences about
#  a population based on a sample. It ensures that each individual or unit in the population has an equal chance of
#  being selected, thus allowing researchers to generalize the findings from the sample to the broader population.

# The Role of Random Sampling in Making Inferences-
# Sampling Distribution and Central Limit Theorem-
# The Central Limit Theorem (CLT) states that if we take a large number of random samples from a population and
# calculate the sample mean for each, the distribution of those sample means will be approximately normal 
# (even if the original population distribution is not). This property allows statisticians to make inferences about
#  population parameters.
# Random sampling allows for the creation of sampling distributions of statistics like the sample mean, which in
# turn helps in the estimation of the population mean and the calculation of confidence intervals.

# Example:
# If you randomly select 30 individuals from a population to calculate the sample mean, the Central Limit 
# Theorem tells you that the sampling distribution of the sample mean will approach a normal distribution as
# the sample size increases, even if the population itself is skewed or has an unusual distribution.

# Estimating Population Parameters-
# When you take a random sample, you can use it to estimate unknown population parameters (such as the population
# mean, variance, or proportions). For example, if you're estimating the average income of all employees at a company,
# random sampling gives you a valid estimate of the population mean income, based on the sample mean.

# Example:
# Suppose you're studying the average age of employees in a large company. You randomly select 50 employees and 
# find that their average age is 35. You can then estimate the average age of all employees in the company, with 
# some level of confidence, by using statistical methods that account for sampling error.

# Hypothesis Testing-
# Random sampling is also essential for hypothesis testing, which allows researchers to test claims or hypotheses 
# about a population based on sample data.
# When you draw a random sample, you can apply statistical tests (e.g., t-tests, chi-squared tests, ANOVA) to test
# the null hypothesis against an alternative hypothesis.

# For example, if a company claims that their employees work an average of 40 hours per week, random sampling would 
# allow you to test this claim using a one-sample t-test to determine whether the observed sample mean differs 
# significantly from 40 hours.




#6- Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

#  Skewness: Definition and Concept -----
#  Skewness refers to the degree of asymmetry or lopsidedness in the distribution of data. In other words, it measures
#  the direction and extent to which a dataset deviates from being symmetrical. A perfectly symmetrical dataset, like
#  a normal distribution, has zero skewness. If the data is skewed, it means that the distribution is not evenly
#  distributed around the mean; instead, it has a tail (longer or shorter) on one side.

#  Skewness is important because it helps to understand the nature of the distribution and indicates whether measures
#  of central tendency (mean, median, mode) are affected by the distribution shape. Skewness can influence the choice
#  of summary statistics and how to interpret data correctly.

# Types of Skewness---
# There are three main types of skewness:

#1- Positive Skew (Right Skew)
#2- Negative Skew (Left Skew)
#3- Zero Skew (Symmetry) 

# How Skewness Affects Data Interpretation---
# Skewness plays a crucial role in understanding the shape of the data distribution, and it directly influences the 
# interpretation of central tendency and dispersion. Here's how skewness affects the data:

#1. Effect on the Mean, Median, and Mode--
#   Positive Skew (Right Skew)----

#  The mean is greater than the median because the long right tail pulls the mean to the right.
#  The mode is typically the smallest, followed by the median, and then the mean.
#  Example: In income data, where a few very high earners exist, the mean income is likely to be higher than the median.

#  Negative Skew (Left Skew)----

# The mean is less than the median because the left tail pulls the mean down.
# The mode is typically the largest, followed by the median, and then the mean.
# Example: In test scores, where most students score highly and a few very low scores exist, the mean test score 
# might be lower than the median.

# Zero Skew (Symmetry)----

# The mean, median, and mode all coincide, and there is no distortion in the central tendency measures.
# Example: In a standard bell curve (normal distribution), the mean, median, and mode are equal.

#2. Influence on Statistical Analysis--
#   Skewed Data and Parametric Tests:

# Many parametric statistical tests, like the t-test, assume that the data follows a normal distribution. If the
# data is highly skewed, these tests may not be appropriate, leading to inaccurate conclusions.
# For skewed data, non-parametric tests (e.g., Mann-Whitney U test, Kruskal-Wallis test) might be more appropriate
# because they do not assume a specific distribution.

# Data Transformation-

# In cases of skewed data, researchers may apply data transformations (e.g., logarithmic, square root, or inverse 
# transformations) to make the data more symmetrical, making it easier to analyze using parametric methods.
# For example, a logarithmic transformation can often help reduce the skewness in data involving large ranges 
# (e.g., incomes or population sizes).

#3. Interpretation of Outliers----
#   Outliers are often more common in skewed distributions:
#   In a positively skewed distribution, outliers are likely to be high values.
#   In a negatively skewed distribution, outliers are likely to be low values.
#   Identifying and understanding outliers in skewed data is crucial because they can disproportionately affect 
#   measures like the mean, leading to misleading conclusions.





#7- What is the interquartile range (IQR), and how is it used to detect outliers?

#  Interquartile Range (IQR): Definition and Concept--
#  The Interquartile Range (IQR) is a statistical measure of the spread or dispersion of a dataset. It represents
#  the range between the first quartile (Q1) and the third quartile (Q3), essentially capturing the middle 50% of
#  the data. The IQR is often used to assess the variability in the dataset and to identify potential outliers.

#  Quartiles and IQR
#  Before diving into the IQR, let's define the quartiles:

#  First Quartile (Q1): The value below which 25% of the data fall. It is the median of the lower half of the dataset.

#  Third Quartile (Q3): The value below which 75% of the data fall. It is the median of the upper half of the dataset.

#  Interquartile Range (IQR): The difference between Q3 and Q1, i.e.  IQR=Q3−Q1

# How the IQR Is Used to Detect Outliers-----

# The IQR is a useful tool for identifying outliers in a dataset. Outliers are data points that significantly differ
# from the majority of the data. Here’s how the IQR is used to detect them:

# Step 1: Calculate the Lower and Upper Boundaries
# To identify outliers, we use the IQR to establish boundaries:

# Lower Bound =  Q1−1.5×IQR
# Upper Bound =  Q3+1.5×IQR

# Step 2: Identify Outliers
# Lower Outliers: Any data point that is below the lower boundary (i.e., smaller than Q1−1.5×IQR) is considered a lower outlier.
# Upper Outliers: Any data point that is above the upper boundary (i.e., greater than Q3+1.5×IQR) is considered an upper outlier.





#8- Discuss the conditions under which the binomial distribution is used.

#  Binomial Distribution: Definition and Conditions-----
#  The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed 
#  number of independent trials of a binary (yes/no) experiment. Each trial results in a success or failure, and the 
#  probability of success is constant across trials.

# For a random variable X that follows a binomial distribution, 𝑋∼Binomial(𝑛,𝑝), where:
# n is the number of trials or experiments,
# p is the probability of success on each trial, and
# (1−p) is the probability of failure.

# Conditions for Using the Binomial Distribution-
# To use the binomial distribution, certain conditions must be satisfied. These conditions ensure that the data 
# can be appropriately modeled by this distribution.

#1. Fixed Number of Trials (n)--
# The number of trials or experiments must be fixed in advance. Each trial is independent of the others.
# Example: Tossing a coin 10 times, rolling a die 5 times, or surveying 100 people.

#2. Two Possible Outcomes per Trial (Binary Outcome)--
# Each trial must have exactly two possible outcomes: success or failure. These outcomes should be mutually exclusive
# and exhaustive, meaning there are no other possibilities.
# Success: The event of interest (e.g., getting a head in a coin toss, passing an exam).
# Failure: The complement of the success (e.g., getting a tail in a coin toss, failing an exam).
# Example: When flipping a coin, the two outcomes are either heads (success) or tails (failure).

#3. Constant Probability of Success (p)
# The probability of success on each trial must be constant. This means that the probability of a success 
# p is the same for every trial.
# if the probability changes between trials, the situation no longer fits the binomial model.
# Example: If the probability of getting a head in a coin toss is always 0.5, the binomial distribution can be used.

#4. Independence of Trials-
# The trials must be independent. This means the outcome of one trial does not affect the outcome of another trial.
# Example: The outcome of one flip of a fair coin does not influence the outcome of another flip.

#5. The Random Variable Should Count the Number of Successes
# The random variable 
# X should represent the number of successes in the 
# n trials. This is a counting process, and the variable of interest is the total number of successes, not individual outcomes.




#9- Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

# Properties of the Normal Distribution-----
# The normal distribution is a continuous probability distribution that is symmetric about its mean. It is one of 
# the most important and widely used distributions in statistics, particularly because many real-world phenomena 
# approximate a normal distribution. The normal distribution is also known as the Gaussian distribution after Carl
# Friedrich Gauss, who first described it.

# Here are the key properties of the normal distribution:

#1. Symmetry---
# The normal distribution is symmetric around its mean (denoted as μ). This means the left and right sides of the
# distribution are mirror images of each other.
# The mean, median, and mode of a normal distribution are all equal and located at the center of the distribution.

#2. Bell-shaped Curve---
# The shape of the normal distribution is bell-shaped, with the highest point at the mean (μ) and tails that extend 
# infinitely in both directions. These tails never touch the horizontal axis, but they get closer and closer as they 
# move farther from the mean.

#3. Defined by Two Parameters: Mean and Standard Deviation---
# The normal distribution is fully defined by its mean (μ) and standard deviation (σ).
# Mean (μ): The central location of the distribution, where the peak occurs.
# Standard deviation (σ): The measure of the spread or width of the distribution. A smaller σ results in a narrow, 
# tall curve, while a larger σ results in a wide, short curve.

#4. The Total Area Under the Curve is 1
# The total area under the normal distribution curve equals 1, which represents the total probability. This is 
# because the normal distribution is a probability distribution, and the area under the curve represents the 
# probability of outcomes.
#5. Asymptotic Nature
# The tails of the normal distribution curve approach the horizontal axis as they extend away from the mean,
# but they never actually touch it. This is known as the asymptotic property.

# The Empirical Rule (68-95-99.7 Rule)----
# The empirical rule is a statistical rule that applies specifically to normal distributions. It provides a quick
# way to estimate the distribution of data in a normal curve, showing how data points are spread in relation to the mean.

# Key Points of the Empirical Rule---
# 68% of the data lies within one standard deviation from the mean:
# In a normal distribution, approximately 68% of all observations fall within the range μ±σ, meaning they are within 
# one standard deviation above or below the mean. 

# 95% of the data lies within two standard deviations from the mean:
# About 95% of all data points fall within the range μ±2σ, meaning they are within two standard deviations above 
# or below the mean.

# 99.7% of the data lies within three standard deviations from the mean:
# Nearly 99.7% of the data points lie within the range μ±3σ, meaning they are within three standard deviations from the mean.
# This rule is particularly helpful in understanding how "spread out" data is in a normal distribution.
#  For example, if you know that the heights of a population are normally distributed with a mean of 5'10" and a 
#  standard deviation of 2 inches, the empirical rule tells you that:

# 68% of people will be between 5'8" and 6'0" (within one standard deviation of the mean).
# 95% of people will be between 5'6" and 6'2" (within two standard deviations of the mean).
# 99.7% of people will be between 5'4" and 6'4" (within three standard deviations of the mean).




#10-  Provide a real-life example of a Poisson process and calculate the probability for a specific event.

#  Real-Life Example of a Poisson Process
#  A Poisson process is a type of stochastic (random) process that models the occurrence of events happening 
#  independently at a constant average rate over time. The key assumption is that the events happen randomly,
#  but with a known average rate of occurrence over a specified interval.

#  A common real-life example of a Poisson process is the number of calls received by a customer service center in 
#  a given period. Suppose that on average, the center receives 10 calls per hour. The number of calls received in a 
#  given hour can be modeled by a Poisson distribution, where the events (calls) occur randomly and independently.

# Example Scenario--
# Situation:
# A customer service center receives an average of 10 calls per hour. We want to calculate the probability of 
# receiving exactly 7 calls in the next hour.

# Step-by-Step Calculation:

# Define the parameters: 
# λ=10 (since on average 10 calls are received per hour),
# k=7 (we want to find the probability of receiving exactly 7 calls),
# The time interval is 1 hour.

#2- Plug the values into the Poisson distribution formula:
#3- Calculate the components:
#4- Substitute these values back into the formula:




#11-  Explain what a random variable is and differentiate between discrete and continuous random variables.

# Random Variable: Definition--
# A random variable is a variable whose values are determined by the outcome of a random experiment or process. 
# In other words, it is a function that assigns a numerical value to each possible outcome in a random event. 
# Random variables are essential in probability theory and statistics because they help model uncertainty and 
# variability in real-world phenomena.

# There are two primary types of random variables: discrete and continuous. Each type has different characteristics
# and is used to model different kinds of data.

# Types of Random Variables---
#1. Discrete Random Variable-

# A discrete random variable takes on a finite or countably infinite set of distinct values. These values can 
# be listed or counted, and there are clear gaps between possible outcomes.

# Characteristics:

# The values of a discrete random variable are countable (either finite or countably infinite).
# The variable can take specific, separate values.
# Each value can be assigned a non-zero probability.
# Examples:

# Number of heads in 10 coin flips: The possible outcomes are 0, 1, 2, 3, ..., up to 10 heads.
# Number of students in a class: This can be any integer value (e.g., 25, 30, 35), but it cannot be a non-integer like 25.5.

#2. Continuous Random Variable--
# A continuous random variable can take on an infinite number of values within a given range or interval. These
# values are not countable, and there are no gaps between possible values. Essentially, continuous random variables
# can take any value within a range, including fractional or decimal values.

# Characteristics:

# The values of a continuous random variable are uncountable and can take any real number within a specified range.
# The variable can take any value in a continuum (e.g., any real number within an interval).

# difference--

# Definition of Discrete Variable-

# A discrete variable is a type of statistical variable that can assume only fixed number of distinct values and 
# lacks an inherent order.

# Also known as a categorical variable, because it has separate, invisible categories. However no values can exist 
# in-between two categories, i.e. it does not attain all the values within the limits of the variable. So, the number
# of permitted values that it can suppose is either finite or countably infinite. Hence if you are able to count the 
# set of items, then the variable is said to be discrete.

# Definition of Continuous Variable--

# Continuous variable, as the name suggest is a random variable that assumes all the possible values in a continuum. 
# Simply put, it can take any value within the given range. So, if a variable can take an infinite and uncountable set
#  of values, then the variable is referred as a continuous variable.
# A continuous variable is one that is defined over an interval of values, meaning that it can suppose any values
# in between the minimum and maximum value. It can be understood as the function for the interval and for each 
# function, the range for the variable may vary.





#12- Provide an example dataset, calculate both covariance and correlation, and interpret the results

# Example Dataset:
# Let’s consider a small dataset of two variables: the number of hours studied and the test scores of 5 students.

# student            Hours Studied (X            Test Score (Y)
#   1                      2                          65
#   2                      3                          70
#   3                      4                          75
#   4                      5                          80
#   5                      6                          85

# Step 1: Calculate Covariance-
# Covariance measures how two variables move together. If both variables increase or decrease together, the covariance
# will be positive; if one increases while the other decreases, the covariance will be negative; and if the variables 
# are independent, the covariance will be close to zero.

#Step 1a: Calculate the Means of X and Y
# Mean of X (Hours Studied):   x= 2+3+4+5+6 / 5 = 4
# Mean of Y (Test Scores):     y= 65+70+75+80+85 / 5 = 75

# Step 1b: Calculate the Covariance

# Now, calculate the terms for each pair of data points:

#Student	          (Xi-X)        (Yi-Y)        (Xi-X)(Yi-Y)
#  1	          2 - 4 = -2	65 - 75 = -10	(-2) * (-10) = 20
#  2	          3 - 4 = -1	70 - 75 = -5	(-1) * (-5) = 5
#  3          	  4 - 4 = 0	    75 - 75 = 0	    (0) * (0) = 0
#  4	          5 - 4 = 1	    80 - 75 = 5	    (1) * (5) = 5
#  5	          6 - 4 = 2	    85 - 75 = 10	(2) * (10) = 20

# Now sum these values: =20+5+0+5+20=50
# ince there are 5 data points, the covariance is: 50 / 5= 10

# Interpretation of Covariance:
# A positive covariance of 10 means that there is a positive relationship between the number of hours studied and 
# the test scores. As the number of hours studied increases, the test scores also tend to increase.


 

```
